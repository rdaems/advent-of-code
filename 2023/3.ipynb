```python
from aocd.models import Puzzle

puzzle = Puzzle(year=2023, day=3)
data = puzzle.input_data.split('\n')
```

```python
data = """467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..""".split('\n')
```

```python
s = len(data)
assert s == len(data[0])

def check(i, start, end):
    number = int(data[i][start:end])
    neighbours = [[i, start-1], [i, end]] + [[i-1, j] for j in range(start-1, end+1)] + [[i+1, j] for j in range(start-1, end+1)]
    for u, v in neighbours:
        if u < 0 or v < 0 or u >= s or v >= s:
            continue
        if data[u][v] != '.':
            return number
    return 0

answer = 0
for i in range(s):
    track = False
    for j in range(s):
        if track and not data[i][j].isdigit():
            answer += check(i, start, j)
            track = False
        if track and j == s - 1:
            answer += check(i, start, j + 1)
        if not track and data[i][j].isdigit():
            track = True
            start = j

puzzle.answer_a = answer
```

```python
number_map = [[-1 for _ in range(s)] for _ in range(s)]
number_map_i = 0
number_map_numbers = []

for i in range(s):
    track = False
    for j in range(s):
        n = data[i][j]
        if n.isdigit():
            number_map[i][j] = number_map_i
        if track and not n.isdigit():
            number = int(data[i][start:j])
            number_map_numbers.append(number)
            number_map_i += 1
            track = False
        elif track and j == s - 1:
            number = int(data[i][start:j+1])
            number_map_numbers.append(number)
            number_map_i += 1
        elif not track and n.isdigit():
            track = True
            start = j

answer = 0
for i in range(s):
    for j in range(s):
        n = data[i][j]
        if n == '*':
            neighbours = [[i-1, j-1], [i-1, j], [i-1, j+1], [i, j-1], [i, j+1], [i+1, j-1], [i+1, j], [i+1, j+1]]
            neighbour_values = []
            for u, v in neighbours:
                if u < 0 or v < 0 or u >= s or v >= s:
                    continue
                if number_map[u][v] >= 0:
                    neighbour_values.append(number_map[u][v])
            neighbour_values = set(neighbour_values)
            if len(neighbour_values) == 2:
                a, b = neighbour_values
                answer += number_map_numbers[a] * number_map_numbers[b]

puzzle.answer_b = answer
```
