```python
from copy import deepcopy
from functools import lru_cache
from itertools import product
from aocd.models import Puzzle
puzzle = Puzzle(year=2021, day=24)
data = puzzle.input_data.split('\n')
```

```python
blocks = [data[18*i:18*i+18] for i in range(14)]

@lru_cache(maxsize=None)
def program(block_i, z, i):
    s = {'w': 0, 'x': 0, 'y': 0, 'z': z}
    for command in blocks[block_i]:
        instruction = command[:3]
        a = command[4]
        b = command[6:]
        if b != '':
            try:
                b = int(b)
            except:
                b = s[b]
        if instruction == 'inp':
            s[a] = i
        elif instruction == 'add':
            s[a] = s[a] + b
        elif instruction == 'mul':
            s[a] = s[a] * b
        elif instruction == 'div':
            s[a] = s[a] // b
        elif instruction == 'mod':
            s[a] = s[a] % b
        elif instruction == 'eql':
            s[a] = int(s[a] == b)
    return s['z']
```

```python
def parse(block):
    alpha = int(block[4][6:])
    beta = int(block[5][6:])
    gamma = int(block[15][6:])
    return alpha, beta, gamma

params = [parse(block) for block in blocks]
for p in params:
    print(p)
@lru_cache(maxsize=None)
def program(block_i, z, i):
    alpha, beta, gamma = params[block_i]
    x = z % 26 + beta
    z = z // alpha
    if x == i:
        return z
    else:
        return 26 * z + i + gamma

def inverse(block_i, z_out):
    solutions = []
    alpha, beta, gamma = params[block_i]
    z_in = 26 * alpha
    i = z_in % 26 + beta
    if 0 < i < 10:
        solutions.append((z_in, i))
    
    
```

```python
def calculate(block_i, z, inputs=[]):
    if block_i == 14:
        print(inputs, z)
        if z == 0:
            return True
        else:
            return False
    if z > bounds[block_i]:
        return False
    for i in range(9, 0, -1):
        new_z = program(block_i, z, i)
        if calculate(block_i + 1, new_z, inputs + [i]):
            return True

calculate(0, 0)
```
