```python
data = '''
3,4,3,1,2
'''
data = data.split('\n')[1:-1]
```

```python
num_fishes = []
fishes = [int(fish) for fish in data[0].split(',')]
for day in range(1, 81):
    for i in range(len(fishes)):
        fishes[i] = fishes[i] - 1
        if fishes[i] < 0:
            fishes.append(8)
            fishes[i] = 6
    num_fishes.append(len(fishes))
    print(f'After {day} days: {len(fishes)}')
# print(f'After {day} days: {len(fishes)} fishes.')
```

```python
# more efficient implementation
fishes = [int(fish) for fish in data[0].split(',')]
num_fishes = [0, 0] + [fishes.count(i) for i in [6, 5, 4, 3, 2, 1, 0]]
for day in range(1, 257):
    num_fishes = [num_fishes[-1]] + [num_fishes[0]] + [num_fishes[1] + num_fishes[-1]] + num_fishes[2:-1]
    print(f'After {day} days: {sum(num_fishes)}')
```
